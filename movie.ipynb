```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on="title")
movies.head(1)
```

```python
# genres
# id
# keyword
# title
# overview
# cast top 5
# crew

movies = movies[["id", "title", "genres","keywords","overview", "cast", "crew"]]
movies.head()
```

```python
movies.isnull().sum()
movies.duplicated().sum()
# movies['overview'].duplicated().sum()
```

```python
from ast import literal_eval
literal_eval(movies.iloc[0]['genres'])
```

```python
def convert(input):
    li = []
    for x in literal_eval(input):
        li.append(x["name"])
    return li

def convert_cast(input):
    li = []
    for i, x in enumerate(literal_eval(input)):
        if i >= 3:
            break
        li.append(x["name"])
    return li
    
def fetch_dir(input):
    li = []
    for x in literal_eval(input):
        if x["job"] == 'Director':
            li.append(x['name'])
            break
    return li
```

```python
# movies['genres'].apply(convert)
# movies.head()
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies['cast'] =  movies["cast"].apply(convert_cast)
movies['crew'] =  movies["crew"].apply(fetch_dir)

movies.head(1)
```

```python
movies.dropna(subset=['overview'], inplace=True)
movies['overview'].isnull().sum()

```

```python
# error occurrs when applying filter to all the datasets, says it is float, so removed one with float
movies['overview'] = movies['overview'].apply(lambda x:x.split() )
```

```python
# movies['genres'].apply(lambda x: i.replace(" ", "") for i in x if isinstance(x, str) else [])
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast']= movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
# movies['tags'] = movies[['overview', 'genres', 'cast', 'crew', 'keywords']]
movies['tags'] = movies['overview'] + movies['genres'] + movies['cast'] + movies['crew'] + movies['keywords']
newdf = movies[['id', 'title', 'tags']]
newdf['tags']
```

```python
newdf['tags'] = newdf['tags'].apply(lambda x: " ".join(x))
newdf['tags'] = newdf['tags'].apply(lambda x: x.lower())
newdf['tags'][0]

# for some reasom when applying lower tags are spaced
```

```python
newdf.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words='english')
```

```python
vectors = cv.fit_transform(newdf['tags']).toarray()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
ps.stem("dancing")
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
stem("in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction samworthington zoesaldana sigourneyweaver jamescameron cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d")
```

```python
newdf['tags'] = newdf['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
# cosine_similarity(vectors).size
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
def recommend(movie):
    index = newdf[newdf['title'] == movie].index[0]
    movies = sorted(list(enumerate(similarity[index])), reverse= True, key=(lambda x:x[1]))[1:6]

    for i in movies:
        print(newdf.iloc[i[0]].title)
    # return [i[0] for i in movies]
recommend('Superman')
```
